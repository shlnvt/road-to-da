## Topik 1: Visualisasi Dasar dengan Matplotlib

```python
import matplotlib.pyplot as plt
import pandas as pd
```

### 1. Anatomi Sebuah Plot
- Figure: Ini adalah kanvas kosong Anda. Seluruh jendela atau halaman tempat Anda akan menggambar.
- Axes: Ini adalah area spesifik di dalam kanvas tempat Anda akan melukis. Sebuah Axes berisi sumbu-x, sumbu-y, dan semua elemen grafik seperti garis, batang, dan titik.

Satu Figure bisa berisi satu atau lebih Axes. Untuk plot sederhana, kita biasanya bekerja dengan satu Figure dan satu Axes.

### 2. Membuat Plot Dasar
#### a. Line Plot (Grafik Garis)

```python
# Membuat Line Plot
plt.figure(figsize=(10, 5)) # Mengatur ukuran kanvas (opsional)
# .index untuk sumbu X, dan kolom 'Sales' untuk sumbu Y
plt.plot(sales_per_year['Year'], sales_per_year['Sales'], marker='o', color='b')

# Menambahkan kustomisasi
plt.title('Tren Penjualan Tahunan (2015-2018)') # Judul Grafik
plt.xlabel('Tahun') # Label Sumbu-X
plt.ylabel('Total Penjualan') # Label Sumbu-Y
plt.grid(True) # Menambahkan grid untuk kemudahan membaca

# Menampilkan plot
plt.show()
```

#### b. Bar Chart (Grafik Batang)

```python
# Membuat Bar Chart
plt.figure(figsize=(8, 6))
plt.bar(sales_per_category['Category'], sales_per_category['Sales'], color=['skyblue', 'salmon', 'lightgreen'])

plt.title('Total Penjualan per Kategori Produk')
plt.xlabel('Kategori Produk')
plt.ylabel('Total Penjualan')
plt.xticks(rotation=15) # Memutar label sumbu-x agar tidak tumpang tindih

plt.show()
```

#### c. Histogram

```python
plt.figure(figsize=(10, 6))
plt.hist(df_cleaned['Sales'], bins=50, color='lightblue', edgecolor='black')
plt.title('Distribusi Nilai Penjualan per Transaksi')
plt.xlabel('Penjualan')
plt.ylabel('Frekuensi (Jumlah Transaksi)')
plt.show()
```

#### d. Scatter Plot (Grafik Sebar)
Digunakan untuk melihat hubungan atau korelasi antara dua variabel numerik. Karena dataset kita hanya memiliki satu kolom numerik utama (Sales) dan tidak ada kolom numerik lain yang relevan untuk dibandingkan (seperti Profit atau Quantity), kita tidak bisa membuat scatter plot yang bermakna.

Tujuan dari scatter plot adalah untuk memplot satu variabel numerik di sumbu-x dan variabel numerik lainnya di sumbu-y untuk melihat apakah ada pola tertentu (misalnya, apakah keduanya cenderung naik bersamaan).

### 3. Kustomisasi Plot
Beberapa kustomisasi paling umum adalah:
- plt.figure(figsize=(lebar, tinggi)): Mengatur ukuran plot.
- plt.title('Judul Anda'): Memberi judul pada plot.
- plt.xlabel('Label Sumbu X'): Memberi label pada sumbu-x.
- plt.ylabel('Label Sumbu Y'): Memberi label pada sumbu-y.
- plt.legend(): Menampilkan legenda.

## Topik 2: Visualisasi Statistik dengan Seaborn

Mengapa menggunakan Seaborn?
- Estetika Superior: Secara default, plot Seaborn terlihat lebih modern dan menarik.
- Sintaks Lebih Sederhana: Untuk plot statistik yang kompleks (seperti box plot, violin plot, heatmap), Seaborn seringkali hanya membutuhkan satu baris kode.
- Integrasi Kuat dengan Pandas: Seaborn dirancang untuk bekerja secara mulus dengan Pandas DataFrame.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Atur tema default Seaborn agar semua plot kita terlihat lebih bagus
sns.set_theme(style="whitegrid")

# --- Memuat Data ---
# Kita akan menggunakan data yang sudah bersih dari modul sebelumnya.
try:
    # Ganti dengan path file yang sudah bersih
    df_cleaned = pd.read_csv(path_file)
    print("--- Data yang sudah bersih berhasil dimuat! ---")
except FileNotFoundError:
    print("File (Cleaned)Dataset Mini Project Modul 2.csv tidak ditemukan.")
    df_cleaned = pd.DataFrame() # Buat DataFrame kosong agar tidak error
```

### 1. Box Plot(Diagram Kotak)

```python
plt.figure(figsize=(12, 7))
sns.boxplot(x='Region', y='Sales', data=df_cleaned)

plt.title('Distribusi Penjualan di Setiap Region')
plt.xlabel('Region')
plt.ylabel('Penjualan')
# Karena nilai sales sangat bervariasi, kita bisa batasi sumbu y untuk melihat detail box-nya
plt.ylim(0, 2000) # Batasi sumbu y dari 0 sampai 2000
plt.show()
```

### 2. Violin Plot

```python
plt.figure(figsize=(12, 7))
sns.violinplot(x='Region', y='Sales', data=df_cleaned)

plt.title('Distribusi Penjualan di Setiap Region (Violin Plot)')
plt.xlabel('Region')
plt.ylabel('Penjualan')
plt.ylim(0, 2000)
plt.show()
```

### 3. Heatmap (Peta Panas)

```python
# Pilih hanya kolom numerik untuk dihitung korelasinya
numerical_cols = df_cleaned.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numerical_cols.corr()

# Membuat Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
# annot=True untuk menampilkan angka berisi korelasi
# cmap='coolwarm' adalah skema warna
# fmt=".2f" untuk format angka 2 desimal

plt.title('Heatmap Korelasi Antar Variabel Numerik')
plt.show()
```

### 4. Pair Plot

```python
# Untuk menghemat wakti, bisa gunakan subset dari data
sns.pairplot(df_cleaned[['Sales', 'Postal Code']]) # Contoh penggunaan
plt.show()
```

## Topik 3: Prinsip Visualisasi Data yang Efektif
### 1. Memilih Jenis Grafik yang Tepat
| Tujuan Analisis | Jenis Grafik yang Direkomendasikan | Contoh Pertanyaan        |
|:---------------:|------------------------------------|----------------------    |
| Perbandingan | Bar Chart (Grafik Batang) | "Region mana yang penjualannya tertinggi?" |
| Tren / Perubahan Waktu | Line Plot (Grafik Garis) | "Bagaimana tren penjualan dari tahun ke tahun?" |
| Distribusi | Histogram, Box Plot, Violin Plot | "Berapa rentang harga penjualan yang paling sering terjadi?" |
| Hubungan / Korelasi | Scatter Plot (Grafik Sebar) | "Apakah ada hubungan antara penjualan dan keuntungan?" |
| Komposisi / Proporsi | Pie Chart, Stacked Bar Chart | "Berapa presentase penjualan dari setiap segmen pelanggan?" |

**Catatan penting tentang Pie Chart:** Gunakan dengan hati-hati! 
Pie chart hanya efektif jika kategorinya sedikit (maksimal 5) dan Anda ingin menunjukkan perbandingan "bagian dari keseluruhan". Untuk perbandingan antar kategori, Bar Chart hampir selalu lebih unggul.

### 2. Menghindari Visualisasi yang Menyesatkan (Misleading Charts)
Sangat mudah untuk secara tidak sengaja (atau sengaja) membuat grafik yang menyesatkan. Berikut adalah beberapa jebakan yang harus dihindari:
- Sumbu Y yang Tidak Dimulai dari Nol: Khususnya untuk Bar Chart, sumbu-y harus selalu dimulai dari 0. Jika tidak, perbedaan antar batang akan terlihat jauh lebih dramatis daripada yang sebenarnya.
[Gambar perbandingan bar chart dengan sumbu Y dari 0 dan tidak dari 0]
- Penggunaan Grafik 3D yang Tidak Perlu: Grafik 3D seringkali mendistorsi perspektif dan membuat perbandingan nilai menjadi lebih sulit. Hindari kecuali benar-benar diperlukan.
- Terlalu Banyak "Tinta" (Chart Clutter): Terlalu banyak warna yang tidak perlu, garis grid yang tebal, atau label yang berlebihan akan mengalihkan perhatian dari data itu sendiri. Buatlah desain yang bersih dan minimalis.
- Korelasi Bukan Berarti Sebab-Akibat: Ingat, hanya karena dua variabel bergerak bersamaan dalam sebuah Scatter Plot, bukan berarti yang satu menyebabkan yang lain. Grafik hanya menunjukkan hubungan, bukan sebab-akibat.

### 3. Teknik Storrytelling with Data
Storytelling with data adalah kemampuan untuk menyajikan data dalam sebuah narasi yang menarik dan persuasif. Tujuannya adalah untuk memandu audiens Anda memahami temuan Anda.
#### a. Pahami Audiens
Siapa yang akan melihat grafik ini? Manajer C-level? Tim teknis? Sesuaikan tingkat detail dan kompleksitas grafik Anda dengan audiens. Manajer mungkin hanya butuh kesimpulan utama, sementara tim teknis mungkin butuh melihat distribusi datanya.
#### b. Miliki Satu Pesan Utama per Grafik
Setiap grafik harus memiliki satu tujuan atau pesan yang jelas. Jangan mencoba memasukkan terlalu banyak informasi ke dalam satu visualisasi. Jika Anda punya dua pesan, lebih baik buat dua grafik.
#### c. Gunakan Teks untuk Memberi Konteks
Jangan biarkan audiens menebak-nebak apa arti grafik Anda. Gunakan elemen teks secara strategis:
- Judul yang Deskriptif: Alih-alih "Penjualan per Region", gunakan judul yang langsung menyatakan insight-nya, seperti "Region West dan East Mendominasi Total Penjualan".
- Anotasi: Beri catatan atau panah langsung pada grafik untuk menyorot titik data yang penting, seperti puncak atau titik terendah yang signifikan.
#### d. Gunakan Warna Secara Strategis
Warna adalah alat yang sangat kuat untuk menarik perhatian.
- Gunakan warna netral (seperti abu-abu) untuk data yang menjadi konteks.
- Gunakan warna yang mencolok (seperti biru atau oranye) untuk menyorot data yang menjadi pesan utama Anda.
