# Proyek Mini Modul 3: Membuat Dashbiard Visual Penjualan Superstore

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Langkah 1: Memuat Data
# Memuat semua data hasil analisis Modul 2
# Membaca dataset yang sudah dibersihkan di modul 2
path_file = 'C:/Users/ASUS/Documents/Road to DA/(Cleaned)Dataset Mini Project Modul 2.csv'
df_cleaned = pd.read_csv(path_file)
print("=============")
print(df_cleaned.head())

# Membaca dataset hasil_sales_per_year.csv
path_file1 = 'C:/Users/ASUS/Documents/Road to DA/hasil_sales_per_year.csv'
sales_per_year = pd.read_csv(path_file1)
print("=============")
print(sales_per_year)

# Membaca dataset hasil_sales_per_category.csv
path_file2 = 'C:/Users/ASUS/Documents/Road to DA/hasil_sales_per_category.csv'
sales_per_category = pd.read_csv(path_file2)
print("=============")
print(sales_per_category)

# Membaca dataset hasil_sales_per_region.csv
path_file3 = 'C:/Users/ASUS/Documents/Road to DA/hasil_sales_per_region.csv'
sales_per_region = pd.read_csv(path_file3)
print("=============")
print(sales_per_region)
```

```python
# Langkah 2: Membuat Layout Dashboard
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(18, 12))

fig.suptitle('Dashboard Analisis Penjualan Superstore (2015 - 2018)', fontsize=20, fontweight='bold')

# Langkah 3: Membuat Setiap Plot
# Plot 1: Total Penjualan per Kategori Produk (Bar Chat)
ax1 = axes[0, 0]
sns.barplot(x='Category', y='Sales', data=sales_per_category, ax=ax1)
ax1.set_title('Penjualan Tertinggi Berdasarkan Kategori', fontsize=14)
ax1.set_xlabel('Kategori Produk', fontsize=12)
ax1.set_ylabel('Total Penjualan', fontsize=12)

# Plot 2: Tren Penjualan Tahunan (Line Plot)
ax2 = axes[0, 1]
ax2.plot(sales_per_year['Year'], sales_per_year['Sales'], marker='o', linestyle='-')
ax2.set_title('Tren Penjualan dari Tahun ke Tahun', fontsize=14)
ax2.set_xlabel('Tahun', fontsize=12)
ax2.set_ylabel('Total Penjualan', fontsize=12)
ax2.grid(True)

# Plot 3: Total Penjualan per Region (Bar Chart dengan Storytelling)
ax3 = axes[1, 0]
colors = ['#1f77b4' if (x == sales_per_region['Sales'].max()) else '#c7c7c7' for x in sales_per_region['Sales']]
sns.barplot(x='Region', y='Sales', data=sales_per_region, palette=colors, ax=ax3)
ax3.set_title(f'Region {sales_per_region['Region'].max()} Mendominasi Penjualan', fontsize=14)
ax3.set_xlabel('Region', fontsize=12)
ax3.set_ylabel('Total Penjualan', fontsize=12)

# Plot 4: Distribusi Penjualan Berdasarkan Segmen Pelanggan (Box Plot)
ax4 = axes[1, 1]
sns.boxplot(x='Segment', y='Sales', data=df_cleaned, ax=ax4)
ax4.set_title('Distribusi Penjualan per Segmen Pelanggan', fontsize=14)
ax4.set_xlabel('Segmen', fontsize=12)
ax4.set_ylabel('Penjualan', fontsize=12)
ax4.set_ylim(0, 1500)

# Langkah 4: Menampilkan Dashboard
plt.tight_layout(rect=[0, 0.03, 1, 0.95])

plt.show()
```
