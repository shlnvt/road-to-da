# Proyek Mini Modul 1: Aplikasi Kasir Sederhana

```python
daftar_produk = [
    {"kode": "A01", "nama": "Buku Tulis", "harga": 5000},
    {"kode": "A02", "nama": "Pensil 2B", "harga": 2500},
    {"kode": "A03", "nama": "Penghapus", "harga": 1500},
    {"kode": "A04", "nama": "Penggaris", "harga": 3000},
    {"kode": "A05", "nama": "Tip-Ex", "harga": 4500},
]

def hitung_dan_cetak_struk(keranjang):
    """
    Fungsi untuk menghitung total belanja dari keranjang,
    memberikan diskon jika memenuhi syarat, memproses pembayaran,
    dan mencetak struk akhir.
    """
 
    subtotal = 0
    print("\n\n=================================")
    print("===== STRUK BELANJA ANDA =====")
    print("=================================")
    
    if not keranjang:
        print("Keranjang Anda kosong.")
        print("=================================")
        return

    for belanjaan in keranjang:
        harga_item = belanjaan['item']['harga']
        kuantitas_item = belanjaan['kuantitas']
        nama_item = belanjaan['item']['nama']
        total_harga_item = harga_item * kuantitas_item
        subtotal += total_harga_item
        print(f"- {nama_item} ({kuantitas_item}x): Rp {total_harga_item}")

    diskon = 0
    if subtotal > 50000:
        diskon = subtotal * 0.10 
    
    total_akhir = subtotal - diskon

    print("---------------------------------")
    print(f"Subtotal     : Rp {int(subtotal)}")
    print(f"Diskon       : Rp {int(diskon)}")
    print("---------------------------------")
    print(f"Total Akhir  : Rp {int(total_akhir)}")
    print("=================================")

    while True:
        try:
            uang_bayar = int(input("Masukkan jumlah uang pembayaran: Rp "))
            if uang_bayar >= total_akhir:
                kembalian = uang_bayar - total_akhir
                print(f"Uang Bayar   : Rp {int(uang_bayar)}")
                print(f"Kembalian    : Rp {int(kembalian)}")
                print("=================================")
                print("Terima kasih telah berbelanja!")
                break
            else:
                print(f"Uang pembayaran kurang. Butuh Rp {int(total_akhir - uang_bayar)} lagi.")
        except ValueError:
            print("Input tidak valid, masukkan angka.")

keranjang_belanja = []

print("\n===== DAFTAR PRODUK =====")
for produk in daftar_produk:
    print(f"Kode: {produk['kode']}, Nama: {produk['nama']}, Harga: {produk['harga']}")

while True:
    
    kode_input = input("Masukkan kode produk (atau ketik 'selesai' untuk selesai): ")
    
    if kode_input.lower() == "selesai":
        break
   
    produk_ditemukan = False
    for produk in daftar_produk:
        if produk['kode'] == kode_input:
            while True:
                try:
                    kuantitas = int(input(f"Masukkan jumlah '{produk['nama']}' yang ingin dibeli: "))
                    if kuantitas > 0:
                        break
                    else:
                        print("Jumlah harus lebih dari 0.")
                except ValueError:
                    print("Input tidak valid, masukkan angka.")

            keranjang_belanja.append({'item': produk, 'kuantitas': kuantitas})
            print(f"-> Berhasil menambahkan {kuantitas} '{produk['nama']}' ke keranjang.")
            produk_ditemukan = True
            break
    
    if not produk_ditemukan:
        print(f"-> Error: Produk dengan kode '{kode_input}' tidak ditemukan.")

hitung_dan_cetak_struk(keranjang_belanja)
```
